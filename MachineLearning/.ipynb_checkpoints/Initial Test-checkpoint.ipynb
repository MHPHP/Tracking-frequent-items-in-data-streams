```python
import numpy as np
import csv
import pandas as pd
import fasttext
import fasttext.util

import torchvision
import matplotlib.pyplot as plt
import random
%matplotlib inline

import torch
from torch.nn.parameter import Parameter
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torch.nn.init as init
```

```python
#queryAndCount er en csv fil lavet i R med ALLe queries fra user-ct-test-collection-01.txt
#og antal af gange hver query optrÃ¦der
df = pd.read_csv("queryAndCount.csv")
print(df.head())
```

```python
max(df["frequency"])
```

```python
# Downloader modellen der laver ord om til vektorer
fasttext.util.download_model('en', if_exists='ignore')
# LÃ¦ser modellen ind.
ft = fasttext.load_model('cc.en.300.bin')
```

```python
#Laver en liste med alle vektorer lavet med fastText. Hver vektor har en lÃ¦ngde af 300 som standard
vector_list = []
for i in range(len(df["query"])):
    vector_list.append(ft.get_word_vector((str(df["query"][i]))))

```

```python
#Laver en vektor for frequency
freq_tensor = torch.tensor(df["frequency"]).float()
```

```python
vector_tensor = torch.tensor(vector_list)
```

```python
len(freq_tensor)
```

```python
#Laver et simple neuralt netvÃ¦rk
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        """"self.inputLayer1=nn.Linear(300, 450, bias = True)
        self.inputLayer13=nn.Linear(450, 600, bias = True)
        self.inputLayer14=nn.Linear(600, 450, bias = True)
        self.inputLayer15=nn.Linear(450, 300, bias = True)"""
        self.inputLayer16=nn.Linear(300, 150, bias = True)
        self.inputLayer2=nn.Linear(150, 75, bias = True)
        self.inputLayer3=nn.Linear(75, 30, bias = True)
        self.inputLayer4=nn.Linear(30, 1, bias = True)
            
    def forward(self,x):

        """x = torch.relu(self.inputLayer1(x))
        x = torch.relu(self.inputLayer13(x))
        x = torch.relu(self.inputLayer14(x))
        x = torch.relu(self.inputLayer15(x))"""
        x = torch.relu(self.inputLayer16(x))
        x = torch.relu(self.inputLayer2(x))
        x = torch.relu(self.inputLayer3(x))
        x = torch.relu(self.inputLayer4(x))
        return torch.squeeze(x)
```

```python

items = 500 # antal af data der bruges. Kan sÃ¦ttes lavt for at gÃ¸re udregninger hurtigere.
testset = torch.utils.data.TensorDataset(vector_tensor[0:items], freq_tensor[0:items])
testloader = torch.utils.data.DataLoader(testset, batch_size=60)
dataiter = iter(testloader)
```

```python
model = Net()
optimizer = torch.optim.Adam(model.parameters(), lr = 0.065)
```

```python
epochs = 25
```

```python
loss_graph = []
criterion = nn.MSELoss()
for epoch in range(epochs):
    train_loss = 0
    for vectors, freq in testloader:
        #print(vectors.size())
        #print(freq.size())
        estimated_freq = model(vectors)
        #print(estimated_freq.size())
        #print(estimated_freq)
        #print(freq)
        loss = criterion(estimated_freq, freq)
        model.zero_grad()
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
    loss_graph.append(train_loss/items)
    print(estimated_freq)
    #print(freq)
    print('Epoch: {} \tTraining Loss: {:.6f}'.format(epoch, loss_graph[epoch]))
```

```python
print(loss_graph)
```
